# Tableaux à deux dimensions (Matrices)


Une **matrice** est une structure de données permettant de stocker des éléments **de même type** sous forme de grille (lignes et colonnes).

$$
A = \begin{bmatrix}
11 & 12 & 13 \\
21 & 22 & 23 \\
31 & 32 & 33 \\
41 & 42 & 43
\end{bmatrix}
$$

La matrice A ci dessus est une matrice 4x3 car elle est formée de **4 lignes** et de **3 colonnes**.

L'accès aux variables se fait par une notation à 2 indices $A[i,j]$ où $i$ est l'indice de la ligne et $j$ l'indice de la colonne.

On a donc $A[2,1]=32$.

```
       0      1     2
    ──────────┬────────► j
  │ ┌─────────┼────────┐
0 │ │  11     12    13 │
  │ ├─────────┼────────┤
1 │ │  21     22    23 │
  │ ├─────────▼────────┤
2 │ │  31   [ 32 ]◄-33-┼─── A[2,1] (Ligne 2, Col 1)
  │ ├──────────────────┤
3 │ │  41     42    43 │
  ▼ └──────────────────┘
  i
```


## Partie algorithme

Jusqu'à présent, nous avons utilisé des tableaux simples (une seule dimension) pour stocker une suite de valeurs. Pour représenter des structures comme un plateau de jeu (échecs, morpion) ou un plan de salle, nous avons besoin de deux coordonnées : la **ligne** et la **colonne**.

**Remarque:**

  Dans la partie algorithmique nous n'étudierons que des matrices de taille fixe (non dynamique) c'est à dire dont le nombre de lignes et de colonnes est défini à l'avance et ne change pas.
   
   *Exemples* : Un plateau de Morpion (3x3), un échiquier (8x8)...

### Déclaration

La syntaxe est très proche de celle des [tableaux](../4_tableaux/tableaux.ipynb)

Voici la syntaxe générale:

```
nom[nombre de lignes, nombre de colonnes]:type

```
**Exemple**:

La matrice $A$ du début de cours se déclarerait donc comme ceci:
```
A[4,3]:entier
```

### Structure et accès

Pour accéder à une "case", on utilise deux indices : le premier désigne la **ligne**, le second la **colonne**.

**Exemple : Jeu de Morpion**
```
    Algorithme : Jeu de Morpion
    Variables:
      - morpion[3,3]:chaine de caractères
    Début
        // Les indices commencent à 0, le centre est donc à l'indice 1
        morpion[1,1] <- "X"
    Fin
```
### Parcours par boucles imbriquées

Pour visiter chaque case d'une matrice, il est nécessaire d'utiliser deux boucles imbriquées. La boucle **externe** parcourt les lignes, et la boucle **interne** parcourt les colonnes de la ligne en cours.

**Exemple : Initialiser une grille 3x3 avec des zéros**
```
    Algorithme : Initialisation
    Variables:
      - grille[3,3]:entier
    Début
        Pour i allant de 0 à 2 (lignes)
            Pour j allant de 0 à 2 (colonnes)
                grille[i,j] <- 0
            Fin Pour
        Fin Pour
    Fin
```


## Implémentation en Python

En Python, on utilise des listes imbriquées (une liste contenant d'autres listes) pour représenter ces structures.

### Cas d'un tableau à taille fixe

Pour créer une grille proprement sans partager les adresses mémoires des lignes (problème de référence), on utilise la compréhension de liste :


````python

# Création d'une grille 3x3 remplie de "-"
plateau = [["-" for j in range(5)] for i in range(3)]

# Accès et modification
plateau[0][1] = "X"   # Modifier les coordonnées pour voir le résultat

# Affichage
for ligne in plateau:
    print(' '.join(ligne))
````


### Cas dynamique (Liste de listes)

On utilise la méthode ``append()`` pour ajouter des lignes ou des éléments.


````python
# Une liste de scores par équipe
scores_equipes = []

# Ajout d'une ligne (sous-liste) pour l'équipe A
scores_equipes.append([12, 15, 10])
# Ajout d'une ligne (sous-liste) pour l'équipe B
scores_equipes.append([10, 12, 8])

# Affichage du résulat:
print("Scores des équipes")
for scores_equipe in scores_equipes:
    print(' '.join([str(score) for score in scores_equipe]))

# Ajout d'un nouveau score:
for scores_equipe in scores_equipes:
    score = int(input("Entrez le nouveau score : "))
    scores_equipe.append(score)

# Affichage du résulat:
print("\nNouvelle grille complétée : ")
for scores_equipe in scores_equipes:
    print(' '.join([str(score) for score in scores_equipe]))
````


### Parcours avec range() et len()

Pour manipuler les indices, on combine les fonctions ``range()`` et ``len()``.


````python
matrice = [[1, 2], [3, 4], [5, 6]]

# Affichage
print('Affichage de la matrice')
for ligne in matrice:
    print(ligne)


print("Affichage des valeurs par coordonnées")
# len(matrice) donne le nombre de lignes (3)
for i in range(len(matrice)):
    # len(matrice[i]) donne le nombre de colonnes de la ligne actuelle (2)
    for j in range(len(matrice[i])):
        print(f"Position [{i}][{j}] = {matrice[i][j]}")
````


<div class="alert alert-info">
    <strong>Note :</strong><br>
    - <code>len(matrice)</code> renvoie le nombre de lignes.<br>
    - <code>len(matrice[0])</code> renvoie le nombre de colonnes.
</div>
